```python
class SimpleHashTable:
  def __init__(self,size):
    self.size = size
    self.table =[None] * size

  def hash_function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.hash_function(key)
    self.table[index] = value

  def search(self, key):
    index = self.hash_function(key)
    return self.table[index]

#Contoh penggunaan
ht = SimpleHashTable(10)
ht.insert("apel", 100)
ht.insert("pisang", 200)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'pisang':", ht.search("pisang"))
```

```python
class LinearProbingHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [None] * size

  def hash_function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.hash_function(key)
    original_index = index

    while self.table[index] is not None:
      if self.table[index][0] == key:
        break
      index = (index + 1) % self.size
      if index == original_index:
        raise Exception("Hash table penuh")

    self.table[index] = (key, value)

  def search(self, key):
    index = self.hash_function(key)
    original_index = index

    while self.table[index] is not None:
      if self.table[index][0] == key:
        return self.table[index][1]
      index = (index + 1) % self.size
      if index == original_index:
        break
    return None

#Contoh penggunaan
ht = LinearProbingHashTable(5)
ht.insert("apel",100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'melon':", ht.search("melon"))
```

```python
class ChainingHashTable:
  def __init__(self, size):
    self.size = size
    self.table = [[] for _ in range (size)]

  def hash_function(self, key):
    return hash(key) % self.size

  def insert(self, key, value):
    index = self.hash_function(key)
    for i, (k,v) in enumerate(self.table[index]):
      if k == key :
        self.table[index][i] = (key, value)
        return
    self.table[index].append ((key, value))

  def search(self, key):
    index = self.hash_function(key)
    for k, v in self.table[index]:
      if k == key:
        return v
    return None

#Contoh penggunaan
ht = ChainingHashTable(5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print ("Nilai untuk 'apel':", ht.search("apel"))
print ("Nilai untuk 'melon':", ht.search("melon"))
```

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        # Menggunakan hash bawaan Python dikombinasikan dengan modulo size
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index

        # Mencari slot kosong secara berurutan jika slot utama sudah terisi
        while self.table[index] is not None:
            if self.table[index][0] == key:
                break
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table penuh")

        self.table[index] = (key, value)

    def search(self, key):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                break
        return None

# EKSPERIMEN SOAL 1
# a. Buat objek Linear Probing HashTable dengan ukuran tabel 5
ht_lap = LinearProbingHashTable(5)

# b. Masukkan pasangan key-value: "A": 10, "B": 20, "C": 30, "D": 40, "E": 50
data_input = {"A": 10, "B": 20, "C": 30, "D": 40, "E": 50}
for k, v in data_input.items():
    ht_lap.insert(k, v)

# c. Cetak isi array self.table setelah semua data dimasukkan
print("HASIL LINEAR PROBING")
print("Isi array self.table final:")
print(ht_lap.table)
```

```python
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        # Membuat list kosong di dalam setiap slot tabel (nested list)
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        # Cek apakah key sudah ada di dalam sub-list untuk diperbarui
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        # Jika belum ada, tambahkan pasangan baru ke akhir sub-list
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

# EKSPERIMEN SOAL 2
# a. Gunakan ukuran tabel 3
ht_chain = ChainingHashTable(3)

# b. Masukkan key-value: "apel": 100, "melon": 200, "lemon": 300, "pisang": 400
data_buah = {"apel": 100, "melon": 200, "lemon": 300, "pisang": 400}
for k, v in data_buah.items():
    ht_chain.insert(k, v)

# c. Cetak isi tabel (self.table)
print("\n=== HASIL SOAL 2 (CHAINING) ===")
print("Isi tabel per slot:")
for idx, slot in enumerate(ht_chain.table):
    print(f"Slot {idx}: {slot}")
```
