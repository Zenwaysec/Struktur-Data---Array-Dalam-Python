```python
def LiniarSearch(arr, N, x):
  for i in range(0, N):
    if (arr[i] == x):
      return i

  return -1

#Driver Code
if __name__ == "__main__":
  arr = [20, 10, 30, 50, 40]
  x = 10
  N = len(arr)

  #function call
  result = LiniarSearch(arr, N, x)
  if(result == -1):
    print("Element yang dicari tidak ada dalam array")
  else:
    print("Element yang dicari berada pada indeks ke=", result)
```

```python
#BINARY SEARCH ITERATIF
# It return location of x in given array arr
def binarySearch(arr, l, r, x):

  while l <= r:
    mid = l + (r - l) // 2

    # check if x is present at mid
    if arr[mid] == x:
      return mid

    # if x is greater, ignore left half
    elif arr[mid] < x:
      l = mid + 1

    # if x is smaller, ignore right half
    else:
      r = mid - 1
  # if we reach here, then the element
  # was not present
  return -1

# Driver code
if __name__ == '__main__':
  arr = [ 2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
  x = 72

  #function call
  result = binarySearch(arr, 0, len(arr)-1, x)
  if result != -1:
    print("Element is present at index", result)
  else:
    print("Element is not present in array")
```

```python
#JUMP SEARCH

import math

def jumpSearch( arr, x, n):

  #finding block size to the jumped
  step = math.sqrt(n)

  #finding the block where element is
  #present (if it is present)

  prev = 0
  while arr[int (min(step,n)-1)] < x:
    prev = step
    step += math.sqrt(n)
    if prev >= n:
      return -1

  #doing a linear search for x in
  #block beginning with prev
  while arr[int(prev)] < x:
    prev += 1

    # if we reached next block or end
    # of array, element is not present
    if prev == min(step, n):
      return -1

  #if element is found
  if arr[int(prev)] == x:
    return prev

  return -1

#Driver code to test function
arr = [ 0, 1, 2, 3, 4, 5, 8, 13, 21, 34, 55, 89, 144, 233, 377, 618]
x = 55
n = len(arr)

#find the index of "x" using jump search
index = jumpSearch(arr, x, n)

#print the index where "x" is located
print("Number", x, "is at index", "%.0f"%index)
```

```python
class Pair:
    def __init__(self):
        self.min = 0
        self.max = 0

def get_min_max(arr, n):
    minmax = Pair()

    # Jika hanya ada satu elemen
    if n == 1:
        minmax.max = arr[0]
        minmax.min = arr[0]
        return minmax

    # Inisialisasi awal untuk dua elemen pertama
    if arr[0] > arr[1]:
        minmax.max = arr[0]
        minmax.min = arr[1]
    else:
        minmax.max = arr[1]
        minmax.min = arr[0]

    # Membandingkan elemen sisanya (dari indeks 2 sampai n-1)
    for i in range(2, n):
        if arr[i] > minmax.max:
            minmax.max = arr[i]
        elif arr[i] < minmax.min:
            minmax.min = arr[i]

    return minmax

# Driver Code
if __name__ == "__main__":
    data_nilai = [1000, 11, 445, 1, 330, 3000]
    ukuran = len(data_nilai)

    hasil = get_min_max(data_nilai, ukuran)
    print("Nilai Minimum adalah:", hasil.min)
    print("Nilai Maksimum adalah:", hasil.max)
```

```python

def pair(arr, x):
    n = len(arr)
    # Inisialisasi awal penunjuk (pointer)
    l = 0
    r = n - 1

    # Langkah 1: Mencari titik pivot (elemen terbesar)
    # Elemen terbesar adalah elemen yang lebih besar dari elemen setelahnya
    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            l = i + 1  # l menjadi indeks elemen terkecil
            r = i      # r menjadi indeks elemen terbesar
            break

    # Langkah 2: Mencari pasangan menggunakan algoritma Two Pointers
    while l != r:
        # Jika jumlah elemen pada posisi l dan r sama dengan target x
        if arr[l] + arr[r] == x:
            print(f"Explanation: Terdapat Pasangan ({arr[l]}, {arr[r]}) dengan jumlah {x}")
            return True

        # Jika jumlah kurang dari target x, geser penunjuk l ke arah nilai yang lebih besar
        elif arr[l] + arr[r] < x:
            l = (l + 1) % n

        # Jika jumlah lebih dari target x, geser penunjuk r ke arah nilai yang lebih kecil
        else:
            r = (n + r - 1) % n

    # Jika tidak ditemukan pasangan setelah seluruh ruang pencarian diperiksa
    print(f"Explanation: Tidak ada pasangan dengan jumlah {x}.")
    return False

# Driver Code untuk menjalankan skenario pengujian
if __name__ == "__main__":
    # Skenario 1
    arr1 = [11, 15, 6, 8, 9, 10]
    x1 = 16
    print(f"Input: arr[] = {arr1}, X = {x1}")
    result1 = pair(arr1, x1)
    print(f"Output: {str(result1).lower()}\n")

    # Skenario 2
    arr2 = [11, 15, 26, 38, 9, 10]
    x2 = 35
    print(f"Input: arr[] = {arr2}, X = {x2}")
    result2 = pair(arr2, x2)
    print(f"Output: {str(result2).lower()}\n")

    # Skenario 3
    x3 = 45
    print(f"Input: arr[] = {arr2}, X = {x3}")
    result3 = pair(arr2, x3)
    print(f"Output: {str(result3).lower()}\n")
```

```python

def cari_elemen_anomali(arr):
    n = len(arr)
    # Membuat array bantuan dengan nilai awal 0
    temp = [0] * n
    angka_terulang = -1
    angka_hilang = -1

    for i in range(n):
        # Menghitung kemunculan setiap angka
        temp[arr[i] - 1] += 1
        if temp[arr[i] - 1] > 1:
            angka_terulang = arr[i]

    for i in range(n):
        # Mencari indeks yang tetap 0
        if temp[i] == 0:
            angka_hilang = i + 1
            break

    print(f"Angka yang terulang adalah: {angka_terulang}")
    print(f"Angka yang hilang adalah: {angka_hilang}")

# Contoh Eksekusi
arr_input = [7, 3, 4, 5, 5, 6, 2]
cari_elemen_anomali(arr_input)
```
