```python
#Bubble Sort
def bubble_sort(arr):
  n = len(arr)
  for i in range (n - 1):
    for j in range (n - 1 - 1):
      if arr[j] > arr [j + 1]:
        arr[j], arr[j + 1] = arr [j + 1], arr[j]
  return arr
#Penggunaan
data = [ 64,34,25,12,22,11,90]
sorted_data = bubble_sort(data.copy())
print("Data sebelum sorting :", data)
print("Data setelah Bubble Sort:", sorted_data)
```

```python
#Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range (n-1):
        min_index = 1
        for j in range (i+1, n):
            if  arr[j] < arr [min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

    #Penerapan
    data = [64,25,12,22,11]
    sorted_data = selection_sort(data.copy())
    print("Data sebelum sorting:", data)
    print("Data sebelum Selection Sort:", sorted_data)
```

```python
#Insertion Sort
def insertion_sort(arr):
  n = len(arr)
  for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key
  return arr
# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

```python
#Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        # Proses penggabungan (merging)
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Memastikan semua elemen sisa dimasukkan
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Merge Sort:", sorted_data)
```

```python
#Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Quick Sort:", sorted_data)
```

```python
# Program Optimasi Harga Marketplace
def quick_sort_harga(produk):
    if len(produk) <= 1:
        return produk

    # Pivot diambil dari harga produk di tengah daftar
    pivot = produk[len(produk) // 2]['harga']

    left = [p for p in produk if p['harga'] < pivot]
    middle = [p for p in produk if p['harga'] == pivot]
    right = [p for p in produk if p['harga'] > pivot]

    return quick_sort_harga(left) + middle + quick_sort_harga(right)

# Data Produk Marketplace
data_produk = [
    {"nama": "Sepatu", "harga": 250000},
    {"nama": "Kaos", "harga": 50000},
    {"nama": "Jaket", "harga": 450000},
    {"nama": "Topi", "harga": 35000},
    {"nama": "Tas", "harga": 150000}
]

print("Daftar Produk Sebelum Diurutkan:")
for p in data_produk:
    print(f"- {p['nama']}: Rp{p['harga']}")

sorted_produk = quick_sort_harga(data_produk)

print("\nDaftar Produk Setelah Diurutkan (Harga Terendah ke Tertinggi):")
for p in sorted_produk:
    print(f"- {p['nama']}: Rp{p['harga']}")
```
