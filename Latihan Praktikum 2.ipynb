```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    # Menambah node di akhir untuk mempermudah pembuatan 3 node awal
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # Penyisipan di tengah (setelah node tertentu)
    def add_in_between(self, prev_node_data, new_data):
        current = self.head
        while current:
            if current.data == prev_node_data:
                new_node = Node(new_data)
                new_node.next = current.next
                new_node.prev = current
                if current.next:
                    current.next.prev = new_node
                current.next = new_node
                return
            current = current.next
        print("Node acuan tidak ditemukan")

    # Hapus dengan kata kunci
    def remove_by_key(self, key):
        current = self.head
        while current:
            if current.data == key:
                if current.prev:
                    current.prev.next = current.next
                else: # Jika yang dihapus adalah head
                    self.head = current.next

                if current.next:
                    current.next.prev = current.prev
                current = None
                return
            current = current.next
        print("Kata kunci tidak ditemukan")

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print("None")


dll = DoubleLinkedList()
dll.append("Mykonos")
dll.append("Velixir")
dll.append("Kahf")
print("Double Linked List awal:")
dll.display()

dll.add_in_between("Mykonos", "Scarlett")
print("\nSetelah penyisipan tengah (setelah Mykonos):")
dll.display()

dll.remove_by_key("Scarlett")
print("\nSetelah hapus dengan kata kunci 'Scarlett':")
dll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Pembuatan 3 node awal
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    # Penyisipan di akhir
    def add_at_end(self, data):
        self.append(data) # Konsepnya sama dengan append pada circular

    # Hapus di akhir
    def remove_end(self):
        if self.head is None:
            return
        if self.head.next == self.head:
            self.head = None
            return

        curr = self.head
        prev = None
        while curr.next != self.head:
            prev = curr
            curr = curr.next
        prev.next = self.head
        curr = None

    def display(self):
        if not self.head:
            print("List Kosong")
            return
        nodes = []
        temp = self.head
        while True:
            nodes.append(str(temp.data))
            temp = temp.next
            if temp == self.head:
                break
        print(" -> ".join(nodes) + " -> (Head)")


cll = CircularLinkedList()
cll.append("Tinky")
cll.append("Dipsy")
cll.append("Lala")
print("Circular Linked List awal:")
cll.display()

cll.add_at_end("Asep")
print("\nSetelah penyisipan di akhir ('Asep')")
cll.display()

cll.remove_end()
print("\nSetelah hapus di akhir:")
cll.display()
```
