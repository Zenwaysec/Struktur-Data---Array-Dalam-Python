```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# [span_1](start_span)Membuat tree secara manual[span_1](end_span)
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# [span_2](start_span)Fungsi inorder traversal untuk mengecek struktur[span_2](end_span)
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Hasil inorder traversal dari binary tree manual:")
inorder(root)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        # [span_5](start_span)Jika node kosong, buat node baru sebagai root/daun[span_5](end_span)
        if root is None:
            return Node(data)

        # [span_6](start_span)Aturan penempatan BST[span_6](end_span)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

# [span_7](start_span)Inisialisasi dan pengujian penyisipan data[span_7](end_span)
bst = BinarySearchTree()
root_bst = None
data_list = [50, 30, 70, 20, 40, 60, 80]

for value in data_list:
    root_bst = bst.insert(root_bst, value)
```

```python
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("Inorder Traversal (Terurut):")
inorder(root_bst)

print("\nPreorder Traversal:")
preorder(root_bst)

print("\nPostorder Traversal:")
postorder(root_bst)
```

```python
def search(node, key):
    # [span_11](start_span)Jika node kosong atau data ditemukan[span_11](end_span)
    if node is None or node.data == key:
        return node

    # [span_12](start_span)Jika kunci lebih kecil, cari ke kiri; jika lebih besar, ke kanan[span_12](end_span)
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

# [span_13](start_span)Contoh pengujian pencarian[span_13](end_span)
key_target = 60
result = search(root_bst, key_target)

if result:
    print(f"\n{key_target} ditemukan dalam tree.")
else:
    print(f"\n{key_target} tidak ditemukan.")
```

```python
# Struktur Tree berdasarkan identitas saya
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi penyisipan otomatis sesuai aturan BST
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

    # Fungsi pencarian elemen
    def search(self, node, key):
        if node is None or node.data == key:
            return node
        if key < node.data:
            return self.search(node.left, key)
        return self.search(node.right, key)

# Traversal Functions
def inorder(node):
    if node:
        inorder(node.left); print(node.data, end=' '); inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' '); preorder(node.left); preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left); postorder(node.right); print(node.data, end=' ')

# Eksekusi Program
bst = BinarySearchTree()
root_bst = None
# Daftar angka unik berdasarkan identitas
data_unik = [25, 21, 2, 82, 97, 9, 1]

for angka in data_unik:
    root_bst = bst.insert(root_bst, angka)

print("--- Hasil Traversal BST ---")
print("Inorder   :", end=' '); inorder(root_bst)
print("\nPreorder  :", end=' '); preorder(root_bst)
print("\nPostorder :", end=' '); postorder(root_bst)

print("\n\n--- Fitur Pencarian ---")
# Cari angka 2 digit terakhir NPM
key1 = 25
res1 = bst.search(root_bst, key1)
print(f"Pencarian {key1}: {'Ditemukan' if res1 else 'Tidak ditemukan'}")

# Cari angka yang tidak ada dalam list
key2 = 100
res2 = bst.search(root_bst, key2)
print(f"Pencarian {key2}: {'Ditemukan' if res2 else 'Tidak ditemukan'}")
```
