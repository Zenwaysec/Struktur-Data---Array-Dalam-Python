```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque ([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end="")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])
#Contoh pemanggilan
bfs(graph, 'A')
```

```python
def find_path(graph, start, end, path= None):
  if path is None:
    path = []

  path = path + [start]
  if start == end:
    return path

  if start not in graph:
    return None

  for neighbor in graph[start]:
    if neighbor not in path:
      newpath = find_path(graph, neighbor, end, path)
      if newpath:
        return newpath
  return None
graph = {
  'A': ['B','C'],
  'B': ['A','B','C'],
  'C': ['A','E'],
  'D': ['B'],
  'E': ['B','F'],
  'F': ['C','E']
}
jalur = find_path(graph, 'A','E')
if jalur:
  print(f"Jalur dari Ake E: {'->'.join(jalur)}")
else:
  print(f"Jalur tidak ditemukan")
```

```python
from collections import deque

# 1. Graf Baru (5 Simpul, 7 Sisi) + Fungsi Tampil
tugas_graph = {
    'V1': ['V2', 'V3', 'V4'],
    'V2': ['V1', 'V3', 'V5'],
    'V3': ['V1', 'V2', 'V4', 'V5'],
    'V4': ['V1', 'V3'],
    'V5': ['V2', 'V3']
}

print("=== SOAL 1: TAMPILAN GRAF ===")
for node, neighbors in tugas_graph.items():
    print(f"{node} -> {neighbors}")

# 2 & 3. Modifikasi BFS (Mengembalikan List Kunjungan)
def bfs_return_list(graph, start):
    visited, queue, order = {start}, deque([start]), []
    while queue:
        vertex = queue.popleft()
        order.append(vertex)
        for n in graph[vertex]:
            if n not in visited:
                visited.add(n)
                queue.append(n)
    return order

# 2 & 4. Fungsi Mencari Jalur (DFS Find Path)
def find_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end: return path
    for neighbor in graph.get(start, []):
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path: return new_path
    return None

# 5. Cek Keterhubungan Graf (DFS + Logika)
def is_connected(graph):
    start = list(graph.keys())[0] if graph else None
    if not start: return True

    visited = set()
    def dfs(node):
        visited.add(node)
        for n in graph[node]:
            if n not in visited: dfs(n)

    dfs(start)
    return len(visited) == len(graph)

# === EKSEKUSI PENGUJIAN SOAL ===
print("\n=== SOAL 2 & 3: HASIL BFS (LIST) ===")
print(bfs_return_list(tugas_graph, 'V1'))

print("\n=== SOAL 2 & 4: JALUR DFS (V1 ke V5) ===")
print(find_path(tugas_graph, 'V1', 'V5'))

print("\n=== SOAL 5: CEK KETERHUBUNGAN ===")
print("Apakah graf tugas terconnected?", is_connected(tugas_graph))
```
